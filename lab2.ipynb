# linear neural network   adaline - adaptive neural network:

```python
import numpy as np 

# adaline adaptive linear neuron
class adaptivelinearneuron(object):

    # niter is the nunmber of iterations
    def __init__(self, rate=0.01,niter=10):
        self.rate=rate
        self.niter=niter
# in each number of iteration(niter) we update the current weights calculates errorbetween y and output and 
# update weight based on the gradient of the cost function 
    def fit(self,x,y):
        # weights 1 is the bias term
        self.weight=np.zeros(1+x.shape[1])
        # no misclassifications
        self.errors=[]
        # cost functions
        self.cost=[]
    
           # Loop through the specified number of iterations
        for i in range(self.niter):
            # Calculate the output by passing the input data through the current model
            output = self.net_input(x)

            # Calculate the errors between target values (y) and predicted values (output)
            errors = y - output

            # Update the weights using the gradient of the cost function with respect to the weights
        #    1: is not includeing the first term which is a bias term so not including the bias term
            # x.T is transposing swappping the rows and columnsso that it can dot product with errors
            
            # x.T.dot(errors) this give the gradient of the cost function with respect to the weights
            # self.rate * x.T.dot(errors) is the step size magnitude of change in weigh
            self.weight[1:] += self.rate * x.T.dot(errors)

            # Update the bias term (the first element in the weight vector)
            self.weight[0] += self.rate * errors.sum()

            # Calculate the cost function (sum of squared errors divided by 2)
            cost = (errors**2).sum() / 2.0

            # Append the cost function value to the list of costs stores cost of each iteration
            self.cost.append(cost)
        return self
    
    def net_input(self,x):
        # Calculate the dot product of the input features and the weights
        return np.dot(x, self.weight[1:]) + self.weight[0]
    
    def activation(self,x):
        # calculate linear activation
        return self.net_input(x)
    
    def predict(self,x):
        # return class label after unit step function np.where creates a newarray with if the
        # condition is true then it will return 1 else -1 and returns the array
        return np.where(self.activation(x) >= 0.0, 1, -1)
    
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df=pd.read_csv('iris.data',header=None)

# selects the first 100 rows and the 5th column 
# The .values attribute returns a NumPy array containing the selected data.
y=df.iloc[0:100,4].values

#  the line converts the target label from string to integer
#  if y is equal to Iris-setosa then return -1 else return 1  cuz adaline model is a numberic 
y=np.where(y=='Iris-setosa',-1,1)

# select the first 100 rows and the 1st and 3rd columns
x=df.iloc[0:100,[0,2]].values

fig ,ax=plt.subplots(nrows=1,ncols=2,figsize=(8,4))

a1n1=adaptivelinearneuron(0.01,10).fit(x,y)

ax[0].plot(range(1,len(a1n1.cost)+1),np.log10(a1n1.cost),marker='o')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('log(Sum-squared-error)')
ax[0].set_title('Adaptive Linear Neuron - Learning rate 0.01')

a1n2=adaptivelinearneuron(0.0001,10).fit(x,y)

ax[1].plot(range(1,len(a1n2.cost)+1),a1n2.cost,marker='o')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Sum-squared-error')
ax[1].set_title('Adaptive Linear Neuron - Learning rate 0.0001')
plt.show()
  
    
```

```python
%matplotlib inline

from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
cm=confusion_matrix(y_true=train_labels,y_pred=rounded_prediction)

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

cm_plot_labels=['no_side_effects','had_side_effects']
plot_confusion_matrix(cm=cm,classes=cm_plot_labels,title='Confusion Matrix')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_batches=ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input).flow_from_directory(directory=train_path,target_size=(224,224),classes=['cat','dog'],batch_size=10)
valid_batches=ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input).flow_from_directory(directory=valid_path,target_size=(224,224),classes=['cat','dog'],batch_size=10)
test_batches=ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input).flow_from_directory(directory=test_path,target_size=(224,224),classes=['cat','dog'],batch_size=10,shuffle=False)
```
