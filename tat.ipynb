```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
```

Number 1-3

```python
data = np.loadtxt('sunspot.txt')
```

```python
filtered_data = data[(data[:, 0] >= 1700) & (data[:, 0] <= 1950)]

print(filtered_data)
```

```python
plt.figure(1)
plt.plot(filtered_data[:, 0], filtered_data[:, 1], 'r-*')

plt.xlabel('Year')
plt.ylabel('Sunspots')
plt.title('Sunspot Activity (1700-1950)')

plt.show()
```

```python
n = 2

L = len(filtered_data)
P = np.array([filtered_data[i:i+n, 1] for i in range(L-n)]).T
T = filtered_data[n:, 1].reshape(-1, 1)

# print('P:', P)
print('T:', T)
print('P shape:', P.shape)
print('T shape:', T.shape)
```

```python
fig = plt.figure(2)
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(P[0, :], P[1, :], T[:, 0], 'bo')

ax.set_xlabel('P[0]')
ax.set_ylabel('P[1]')
ax.set_zlabel('T')
ax.set_title('Input and Output Matrices')

plt.show()
```

```python
print(P)

P = P.transpose()

print(P)

# print(P)
Pu, _, Tu, _ = train_test_split(P, T, train_size=200, shuffle=False)
# print('Pu:', Pu)
# print('Tu:', Tu)
# print('Pu shape:', Pu.shape)
# print('Tu shape:', Tu.shape)
```

```python
net = MLPRegressor(hidden_layer_sizes=(1,), activation='identity', solver='lbfgs')

net.fit(Pu, Tu)

print('Weight coefficients:', net.coefs_)
```

```python
print('Neuron weight coefficient values:')
print('w1:', net.coefs_[0][0][0])
print('w2:', net.coefs_[0][1][0])
print('b:', net.intercepts_[0][0])
w1 = net.coefs_[0][0][0]
w2 = net.coefs_[0][1][0]
b = net.intercepts_[0][0]
```

```python
Tsu = net.predict(Pu)
plt.plot(Tu, label='True values')
plt.plot(Tsu, label='Forecasted values')
plt.legend()
plt.show()
```

```python
Ts = net.predict(P)

plt.plot(T, label='True values')
plt.plot(Ts, label='Forecasted values')
plt.legend()
plt.show()
```

```python
e = Ts - T

plt.plot(e)
plt.title('Forecast error')
plt.xlabel('Time')
plt.ylabel('Error')
plt.show()
```

```python
plt.hist(e, bins=20)
plt.title('Forecast error histogram')
plt.xlabel('Error')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the Mean-Square-Error (MSE)
MSE = np.mean((Ts - T)**2)

# Calculate the Mean Absolute Deviation (MAD)
MAD = np.mean(np.abs(Ts - T))

print(f'Mean-Square-Error (MSE): {MSE:.2f}')
print(f'Mean Absolute Deviation (MAD): {MAD:.2f}')
```
